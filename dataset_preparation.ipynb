This notebook is the dataset preprocessing part of the project

```python
import os
```

```python
def rename_files_in_folder(folder_path):
    """
    Renames files in a folder with the folder name plus their position.

    Args:
        folder_path (str): The path to the folder.
    """
    files = os.listdir(folder_path)
    for i, file in enumerate(files):
        if os.path.isfile(os.path.join(folder_path, file)):
            new_name = f"{os.path.basename(folder_path)}_{i+1}{os.path.splitext(file)[1]}"
            try:
                os.rename(os.path.join(folder_path, file), os.path.join(folder_path, new_name))
            except FileExistsError:
                print(f"Error: File '{new_name}' already exists in the folder. Skipping.")

# Example usage:
folder_to_rename = 'Multiple_Class_Eras_Tour/Acoustic'
rename_files_in_folder(folder_to_rename)
folder_to_rename = 'Multiple_Class_Eras_Tour/Fearless'
rename_files_in_folder(folder_to_rename)
folder_to_rename = 'Multiple_Class_Eras_Tour/Folkmore'
rename_files_in_folder(folder_to_rename)
folder_to_rename = 'Multiple_Class_Eras_Tour/Lover'
rename_files_in_folder(folder_to_rename)
folder_to_rename = 'Multiple_Class_Eras_Tour/Midnights'
rename_files_in_folder(folder_to_rename)
folder_to_rename = 'Multiple_Class_Eras_Tour/Reputation'
rename_files_in_folder(folder_to_rename)
folder_to_rename = 'Multiple_Class_Eras_Tour/Speak_Now'
rename_files_in_folder(folder_to_rename)
```
